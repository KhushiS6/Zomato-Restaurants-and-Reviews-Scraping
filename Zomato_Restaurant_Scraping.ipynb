```python
# Importing required libraries
import pandas as pd
from bs4 import BeautifulSoup
import selenium
import requests
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Navigating to the Zomato website, targeting a specific region (Mumbai)
driver = webdriver.Chrome()
wait = WebDriverWait(driver,10,poll_frequency=1)
driver.get('https://www.zomato.com/mumbai/delivery')
time.sleep(5)
driver.maximize_window()
```

```python
# Creating a pandas dataframe to store the extracted data
df = pd.DataFrame()
```

```python
# Scrapping restaurants names using selenium and required tag
elements = driver.find_elements(By.CSS_SELECTOR, 'h4.sc-1hp8d8a-0')
name=[]
for element in elements:
    name.append(element.text)
print(name)

# Checking the count of the scrapped restaurants names
print(len(name))
```

```python
# Scrapping restaurants ratings using selenium and required tag
elements = driver.find_elements(By.CSS_SELECTOR, 'div.sc-1q7bklc-1')
rating=[]
for element in elements:
    rating.append(element.text)
print(rating)

# Checking the count of the scrapped restaurants ratings
print(len(rating))
```

```python
# Scrapping restaurants cuisines and costs for one using selenium and required tag
elements = driver.find_elements(By.CSS_SELECTOR, 'div.sc-gggouf')
cuisine=[]
for element in elements[1::2]:
    cuisine.append(element.text)
print(cuisine)
```

```python
# Checking the count of the scrapped restaurants cuisines
print(len(cuisine))
```

```python
# Adding the scrapped data to the created dataframe
zomato={"Retaurant_Name":name,"Rating":rating,"Cuisine_Type":cuisine}
df=pd.DataFrame(zomato)
df
```

```python
# Splitting the scrapped cuisines and costs for one data into separate columns
df[["Cuisine_Type","Cost for One"]]=df["Cuisine_Type"].str.split("\n",expand=True)
```

```python
df
```

```python
# Scrapping restaurants delivery time using selenium and required tag
elements = driver.find_elements(By.CSS_SELECTOR, 'div.min-basic-info-right')
delivtime=[]
for element in elements:
    delivtime.append(element.text)
print(delivtime)
```

```python
# Checking the count of the scrapped restaurants delievry time
print(len(delivtime))
```

```python
# Adding the scrapped data to the created dataframe
df['Delivery_Time'] = delivtime
df
```

```python
# Scrapping restaurants offers(if any) using selenium and required tag
elements = driver.find_elements(By.CSS_SELECTOR, 'div.sc-ZUflv.jYdpIN')
offer=[]
for element in elements:
    offer.append(element.text)
print(offer)

# Checking the count of the scrapped restaurants offers
print(len(offer))
```

```python
# Adding the scrapped data to the created dataframe and filling the blank rows
df['%_Off_Offers'] = offer
df['%_Off_Offers'].replace('','NA', inplace=True)
df
```

```python
# Scrapping restaurants urls using selenium and required tag
elements = driver.find_elements(By.CSS_SELECTOR, 'a.sc-btewqU.dlikcC')
url_link=[]
for element in elements:
    link = element.get_attribute('href')
    url_link.append(link)
print(url_link)

# Checking the count of the scrapped restaurants url
print(len(url_link))
```

```python
# Adding the scrapped data to the created dataframe
df['URL_Link'] = url_link
df
```

```python
# Changing the data type of the dataframe to the required type
df['Rating'] = df['Rating'].astype('float')
df.info()
```

```python
# Exporting the dataframe to a excel file for sample output
df.to_excel('restaurants.xlsx', index=False)
```
