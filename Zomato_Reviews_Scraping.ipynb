```python
# Importing required libraries
import pandas as pd
from bs4 import BeautifulSoup
import selenium
import requests
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Navigating to the Zomato website, targeting a specific region (Mumbai), for use with Selenium
driver = webdriver.Chrome()
wait = WebDriverWait(driver,10,poll_frequency=1)
driver.get('https://www.zomato.com/mumbai/burger-king-dadar-west/reviews')
time.sleep(5)
driver.maximize_window()
```

```python
# Requesting and scrapping the website data, targeting a specific region (Mumbai), for use with BeautifulSoup
from urllib.request import Request, urlopen

css=[]
for i in range(1,11):
    req = Request(
        url='https://www.zomato.com/mumbai/burger-king-dadar-west/reviews?page='+str(i)+'&sort=dd&filter=reviews-dd', 
        headers={'User-Agent': 'Mozilla/5.0'}
    )
    webpage = urlopen(req).read()
    css.append(webpage)
data = str(css)
data
```

```python
# Storing the data in soup
soup = BeautifulSoup(data, 'html.parser')
```

```python
# Transforming the required data and storing them into a list
d = soup.find_all('p', class_= 'sc-1hez2tp-0')
l=[]
for i in d:
    l.append(i.text)
    
print(l)
```

```python
# Cleaning and filtering the required data from the list for use
new_l=[]
for i in l:
    if i != 'Detect current location' and i != 'Using GPS' and i != 'View Gallery' and i != 'All Reviews' and i != 'Newest First':
        new_l.append(i)
        
print(new_l)
print(len(new_l))
```

```python
# Getting user names by filtering the required data from the list
name = []
for n in new_l[0::4]:
    name.append(n)
    
print(name)
print(len(name))
```

```python
# Getting time of posting the review by filtering the required data from the list
time = []
for n in new_l[1::4]:
    time.append(n)
print(time)
print(len(time))
```

```python
# Getting user review by filtering the required data and handling nulls from the list
rev_text = []
for n in new_l[2::4]:
    if n=='':
        n='NA'
    rev_text.append(n)
    
print(rev_text)
print(len(rev_text))
```

```python
# Getting nummber of votes and comments by filtering the required data from the list
vote_comment = []
for n in new_l[3::4]:
    vote_comment.append(n)
    
print(vote_comment)
print(len(vote_comment))
```

```python
# Scrapping the count of total reviews by the user using selenium and required tag
reviews = []
pg = 6
for i in range(1, 11):
    # Wait for elements to be present
    elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'span.sc-ljUfdc'))
    )
    
    for element in elements[::2]:
        reviews.append(element.text)

    # Wait for the next page link to be clickable
    next_page = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="root"]/div/main/div/section[4]/div/div/section/div[3]/div[2]/div/a[' + str(pg) + ']'))
    )
    next_page.click()
    
    # Update the page number
    if pg < 7:
        pg += 1

# Print the total count of elements
print(reviews)
print(len(reviews))
```

```python
# Scrapping the count of followers of the user using selenium and required tag
followers = []
pg = 6
for i in range(1, 11):
    # Wait for elements to be present
    elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'span.sc-ljUfdc'))
    )
    
    for element in elements[1::2]:
        followers.append(element.text)

    # Wait for the next page link to be clickable
    next_page = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="root"]/div/main/div/section[4]/div/div/section/div[3]/div[2]/div/a[' + str(pg) + ']'))
    )
    next_page.click()
    
    # Update the page number
    if pg < 7:
        pg += 1

# Print the total count of elements
print(followers)
print(len(followers))
```
